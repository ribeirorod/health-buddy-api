```python
def getNutrients(x):
    results = []
    j={}
    for i in x:
        id = i['nutrient']['id']
        j['nutrient_name'] = i['nutrient']['name']
        j['nutrient_unit'] = i['nutrient']['unitName']
        j['nutrient_amount'] = i['amount']
        results.append({id : j})
        j={}
    return results

df['foodNutrients'] = df['foodNutrients'].head(15).apply(getNutrients)
```
