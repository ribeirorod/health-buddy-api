```python
import pandas as pd
import json
```

```python

with open('sample.json', 'r') as j:
     contents = json.loads(j.read())
df = pd.json_normalize(contents)
```

```python
df.columns
```

```python
df.drop(columns=['foodClass','foodAttributes','modifiedDate','availableDate','dataSource','dataType','foodUpdateLog'], inplace=True)
df.head(5)
```

```python
df['foodNutrients'][0]
```

```python
def getNutrients(x):
    results = []
    j={}
    for i in x:
        id = i['nutrient']['id']
        j['nutrient_name'] = i['nutrient']['name']
        j['nutrient_unit'] = i['nutrient']['unitName']
        j['nutrient_amount'] = i['amount']
        results.append({id : j})
        j={}
    return results

df['foodNutrients'] = df['foodNutrients'].apply(getNutrients)
```
